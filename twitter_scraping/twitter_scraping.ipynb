```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome(service= ChromeService (ChromeDriverManager().install()))
```

```python
url = "https://twitter.com/businessbarista/status/1714724312798593060"
driver.get(url)
time.sleep(5)
main_row_list= []

previous_height = driver.execute_script('return document.body.scrollHeight' )

count = 0
while (True):
    main_article_divs = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div/div[2]/main/div/div/div/div/div/section/div/div/div")
    print("Page length = ", len(main_article_divs))
    time.sleep(1)

    for each_article_div in main_article_divs:
        temp_row_list = []
        try:
            print("Good")
            full_name = each_article_div.find_element(By.XPATH, "div/div/article/div/div/div[2]/div[2]/div[1]/div[1]/div[1]/div/div/div[1]/div/a/div/div[1]/span/span").text
            username = each_article_div.find_element(By.XPATH, "div/div/article/div/div/div[2]/div[2]/div[1]/div[1]/div[1]/div/div/div[2]/div/div[1]/a/div/span").text
            comment =  each_article_div.find_element(By.XPATH, "div/div/article/div/div/div[2]/div[2]/div[2]/div/span").text
            temp_row_list.append(full_name)
            temp_row_list.append(username)
            temp_row_list.append(comment)
            main_row_list.append(temp_row_list)
        except NoSuchElementException:
            print("NoSuchElementException")
        else:
            print("Something went wrong")
    
    
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight); ' )
    time.sleep(4)

    new_height =  driver.execute_script('return document.body.scrollHeight')
    if(previous_height == new_height):
        break
    previous_height = new_height
```
